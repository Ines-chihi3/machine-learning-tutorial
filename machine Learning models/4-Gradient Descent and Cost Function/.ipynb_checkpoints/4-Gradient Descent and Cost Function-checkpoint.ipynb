```python
import numpy as np

def gradient_descent(x,y):
    m_curr = b_curr = 0  # start with 0 and then baby steps 
    iterations = 10000 #nbr d'etiteration , steps 
    n = len(x)
    learning_rate = 0.08
    y_predicted = m_curr * x + b_curr
    mincost=sum([val**2 for val in (y-y_predicted)])
    for i in range(iterations):  #etirate 
        y_predicted = m_curr * x + b_curr
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])   #cost fucntion
        md = -(2/n)*sum(x*(y-y_predicted))   #md
        bd = -(2/n)*sum(y-y_predicted)  #bd
        m_curr = m_curr - learning_rate * md  #next m formu
        b_curr = b_curr - learning_rate * bd  #next b
        print ("m {}, b {}, cost {} iteration {}, mincost {},".format(m_curr,b_curr,cost, i,mincost))
        if cost <mincost :
            mincost=cost ;
    print (mincost)
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

gradient_descent(x,y)
```
